```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer("clip-ViT-B-32")
```

```python
img01 = "/content/01.09.26.jpeg"
img01 = Image.open(img01)
```

```python
plt.figure(figsize = (6,6))
plt.imshow(img01)
plt.axis("off")
plt.title("Image for similarity Comparison")
plt.show()
```

```python
desc = ["Two dogs in the snow","A cat on a table","A happy dog play fetch",
        "A dog sleeping on a couch","Two cats are fighting in a garden","An icecream seller on road","taj mehal on a view"]
```

```python
desc_emb = model.encode(desc)
img_emb = model.encode(img01)
```

```python
sim_score = np.dot(desc_emb,img_emb)
best_score = sim_score.max()
```

```python
best_desc = desc[np.argmax(sim_score)]
```

```python
print(f"\n Best match for the image is : '{best_desc}'with score of {best_score:.4}")

plt.figure(figsize = (6,6))
plt.imshow(img01)
plt.axis("off")
plt.title("Image for similarity Comparison")
plt.show()
```
